## ì‚¬ìš©ì ì •ì˜ í•¨ìˆ˜ ë§Œë“¤ê¸° (ê¸°ë³¸)

ë‹¤ìŒì˜ ë‹¤ì„¯ í•¨ìˆ˜ë¥¼ ë§Œë“­ë‹ˆë‹¤.<br>

- sum(a,b): aì™€ bê°’ì„ ë”í•œ ê°’ì„ ì¶œë ¥ê°’ìœ¼ë¡œ ëŒë ¤ì¤Œ<br>
- multiply(a,b): aì™€ bê°’ì„ ê³±í•œ ê°’ì„ ì¶œë ¥ê°’ìœ¼ë¡œ ëŒë ¤ì¤Œ<br>
- subtract(a,b): aì—ì„œ bê°’ì„ ëº€ ê°’ì„ ì¶œë ¥ê°’ìœ¼ë¡œ ëŒë ¤ì¤Œ<br>
- multiply_three_numbers(a,b,c): $a*b*c$ë¥¼ ëª¨ë‘ ê³±í•œ ê°’ì„ ì¶œë ¥ê°’ìœ¼ë¡œ ëŒë ¤ì¤Œ<br>
- add_and_multiply(a,b,c): $(a+b)*c$ë¥¼ ê³„ì‚°í•œ ê°’ì„ ì¶œë ¥ê°’ìœ¼ë¡œ ëŒë ¤ì¤Œ<br>

ì„ì‹œ ë³€ìˆ˜ x, y, zë¥¼ ë§Œë“¤ì–´ì„œ ê°ê°ì— ì ì ˆí•œ ê°’ì„ ëŒ€ì…í•œ í›„,<br>
ìœ„ì—ì„œ ì •ì˜í•œ ì‚¬ìš©ì í•¨ìˆ˜ë“¤ì„ í•˜ë‚˜ í•˜ë‚˜ í˜¸ì¶œí•˜ê³  ë‚œí›„, ê²°ê³¼ê°’ì„ ``print()`` ë¬¸ë²•ì„ ì‚¬ìš©í•˜ì—¬ ì¶œë ¥í•˜ë„ë¡ í•©ë‹ˆë‹¤.

````python
def sum(x,y):
    return x+y

def multiply(x,y):
    return x*y

def substract(x,y):
    return x-y

def multiply_three_numbers(x, y, z):
    return x*y*z

def add_and_multiply(x, y, z):
    return (x+y)*z

print(sum(3,1))
print(multiply(3,1))
print(substract(3,1))
print(multiply_three_numbers(3,1,2))
print(add_and_multiply(3,1,2))
````

````python
assert sum(1,2) == 3
assert multiply(1, 2) == 2
assert subtract(1, 2) == -1
assert multiply_three_numbers(1, 2, 3) == 6
assert add_and_multiply(1, 2, 3) == 9
````

## ì‚¬ìš©ì ì •ì˜ í•¨ìˆ˜ ë§Œë“¤ê¸° (ì‹¬í™”1)

## ì‚¬ìš©ì ì •ì˜ í•¨ìˆ˜ ë§Œë“¤ê¸° (ì‹¬í™”2)
